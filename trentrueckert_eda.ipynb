#### **Imports**

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

#### **Step 1. Data Acquisition**

```python
# Load the Iris dataset into DataFrame
df = sns.load_dataset('iris')

# Inspect first rows of the DataFrame
print(df.head())
```

#### **Step 2. Initial Data Inspection**

```python
# Display the first 10 rows of the dataframe
print(df.head(10))
```

```python
# Check the shape of the dataframe
print(df.shape)
```

```python
# Display the data types of each column of the dataframe
print(df.dtypes)
```

#### **Step 3. Initial Descriptive Statistics**

```python
# Display summary statistics for each column of the dataframe
print(df.describe())
```

#### **Step 4. Initial Data Distribution for Numerical Columns**

```python
# Histograms for a column and all columns

df['petal_length'].hist()

df.hist()

plt.show()
```

The shapes of the histograms are all fairly different. The sepal_width shape is roughly symmetrical, the sepal_length is as well (but a little more variation from the typical bell curve), the petal_length and petal_width are both skewed right. Some of the petal lengths and widths are fairly small while most of the others are around 5, and the sepal lengths and widths vary quite a bit more.

#### **Step 5. Initial Data Distribution for Categorical Columns**

```python
# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

There are 50 flowers of each species, an even amount for each one from the 150 total.

#### **Step 6. Initial Data Transformation and Feature Engineering**

```python
# Renaming sepal_length column
df.rename(columns={'sepal_length': 'Sepal_Length'}, inplace=True)

# Renaming sepal_width column
df.rename(columns={'sepal_width': 'Sepal_Width'}, inplace=True)

# Renaming petal_length column
df.rename(columns={'petal_length': 'Petal_Length'}, inplace=True)

# Renaming petal_width column
df.rename(columns={'petal_width': 'Petal_Width'}, inplace=True)

# Renaming species column
df.rename(columns={'species': 'Species'}, inplace=True)
```

```python
# Adding a new column
df['Sepal_Area'] = df['Sepal_Length'] * df['Sepal_Width']

# Adding a new column
df['Petal_Area'] = df['Petal_Length'] * df['Petal_Width']
```

#### **Step 7. Initial Visualizations**

```python
sns.pairplot(df, hue='Species')
plt.show()
```

The pairplot function is very interesting because it shows every combination of our columns with each different type of flower. From this, it looks like the versicolor and virginica are very similar, but the setosa looks to be very different from them. 

```python
sns.relplot(data=df, x="Petal_Length", y="Petal_Width", hue="Species")
plt.show()
```

The petals of the setosa are much smaller in length and width than the other two species.

```python
sns.boxplot(data=df, x="Sepal_Length", y="Sepal_Width")
plt.show()
```

There is an extremely wide range of sepal lengths vs sepal widths for all of the species.

```python
sns.jointplot(data=df, x="Petal_Width", y="Petal_Length", kind="scatter")
plt.show()
```

Petal length and petal width have a positive correlation, so as length increases, so does width.
